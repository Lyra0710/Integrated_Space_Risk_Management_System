```python
import os
import csv
```

```python
file_path = 'vizier_votable.tsv'
start_line = 142
end_line = 340
lines_to_read = []

# Read .tsv as a .txt
with open(file_path, 'r') as file:
    for line_number, line in enumerate(file, start=1):
        if start_line <= line_number <= end_line:
            lines_to_read.append(line)
        elif line_number > end_line:
            break

lines_to_read = [i[:-2] if i.endswith('\n') else i for i in lines_to_read]
for i in range(0,10):
    print(lines_to_read[i])
```

```python
# Manipulating the list
header_temp = lines_to_read[0].split('\t') + lines_to_read[1].split('\t')
print(header_temp)
headers = []
for i in header_temp:
    headers += (i.split(';'))

print(headers)
```

```python
# with open('vizier_votable.csv', mode='w', newline='') as file:
#     writer = csv.writer(file)
#     writer.writerow(headers)
```

```python
# To get the body_temp
body_temp = [i.strip() for i in lines_to_read[3:]]
print(body_temp[0:5])
```

```python
body = []
for i in body_temp:
    body.append(list(i.split(';')))

for i in range(0,5):
    print(body[i])
```

```python
with open('vizier_votable1.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(headers)
    writer.writerows(body)
```
